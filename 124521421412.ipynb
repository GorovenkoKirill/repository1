```sql
/****** Object:  UserDefinedFunction [dbo].[LevenshteinDistance]    Script Date: 14.05.2025 13:37:00 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- Добавляем эту функцию в базу данных MedBD
CREATE FUNCTION [dbo].[LevenshteinDistance](@s1 NVARCHAR(MAX), @s2 NVARCHAR(MAX))
RETURNS INT
AS
BEGIN
    DECLARE @s1_len INT, @s2_len INT, @i INT, @j INT, @c INT, @c_temp INT, @cost INT
    DECLARE @dist TABLE (i INT, j INT, cost INT)
    
    SET @s1_len = LEN(@s1)
    SET @s2_len = LEN(@s2)
    
    IF @s1_len = 0 RETURN @s2_len
    IF @s2_len = 0 RETURN @s1_len
    
    -- Инициализация первой строки и первого столбца
    SET @i = 0
    WHILE @i <= @s1_len
    BEGIN
        INSERT INTO @dist (i, j, cost) VALUES (@i, 0, @i)
        SET @i = @i + 1
    END
    
    SET @j = 1
    WHILE @j <= @s2_len
    BEGIN
        INSERT INTO @dist (i, j, cost) VALUES (0, @j, @j)
        SET @j = @j + 1
    END
    
    -- Заполнение матрицы
    SET @i = 1
    WHILE @i <= @s1_len
    BEGIN
        SET @j = 1
        WHILE @j <= @s2_len
        BEGIN
            -- Стоимость замены
            IF SUBSTRING(@s1, @i, 1) = SUBSTRING(@s2, @j, 1)
                SET @cost = 0
            ELSE
                SET @cost = 1
                
            -- Получаем минимальную стоимость из трех возможных операций
            SELECT @c_temp = cost FROM @dist WHERE i = @i-1 AND j = @j
            SET @c = @c_temp + 1 -- Удаление
            
            SELECT @c_temp = cost FROM @dist WHERE i = @i AND j = @j-1
            IF @c > (@c_temp + 1) SET @c = @c_temp + 1 -- Вставка
            
            SELECT @c_temp = cost FROM @dist WHERE i = @i-1 AND j = @j-1
            IF @c > (@c_temp + @cost) SET @c = @c_temp + @cost -- Замена
            
            -- Сохраняем результат
            INSERT INTO @dist (i, j, cost) VALUES (@i, @j, @c)
            
            SET @j = @j + 1
        END
        SET @i = @i + 1
    END
    
    -- Возвращаем результат (правый нижний угол матрицы)
    SELECT @c = cost FROM @dist WHERE i = @s1_len AND j = @s2_len
    RETURN @c
END
GO
```
